```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# importing the dataset
dataset = pd.read_csv('Position_Salaries.csv')
```

```python
# iloc- interger location to spilit dependant and indep. variable 
X = dataset.iloc[:,1:2].values
Y = dataset.iloc[:,2].values
```

```python
X
```

```python
#Fitting Linear Regression to the dataset
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X,Y)
```

```python
#Fitting Polynomial Regression to the dataset
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=4) # will also add columns of 1's is entered into our matrics to include constant b0 into our matrics
X_poly = poly_reg.fit_transform(X) # will add x^4 values to the polynomial

lin_reg_2 = LinearRegression()# to fit to new generated matrics (X_poly).
lin_reg_2.fit(X_poly,Y)
```

```python
#visualing the Linear Regression results
plt.scatter(X,Y, color ='red')
plt.plot(X, lin_reg.predict(X), color= 'blue')# in place of Y, we are entering the values predicted by linear regression.
plt.title('Truth or Bluff(Linear Regression)')
plt.xlabel('Postion Level')
plt.ylabel('Salary')
plt.show()
```

```python
#visualing the PolynomialRegression results
plt.scatter(X,Y, color ='red')
plt.plot(X, lin_reg_2.predict(poly_reg.fit_transform(X)), color= 'blue')  # where poly_reg.fit_transform(X)= X_poly # in place of Y, we are entering the values predicted by polynomial regression.
plt.title('Truth or Bluff(Polynomial Regression)')
plt.xlabel('Postion Level')
plt.ylabel('Salary')
plt.show()
```

```python
#visualing the PolynomialRegression results and  removing straight lines in the curve
X_grid = np.arange(min(X),max(X),0.1)
X_grid = X_grid.reshape(len(X_grid),1)
plt.scatter(X,Y, color ='red')
plt.plot(X_grid, lin_reg_2.predict(poly_reg.fit_transform(X_grid)), color= 'blue')  
plt.title('Truth or Bluff(Polynomial Regression)')
plt.xlabel('Postion Level')
plt.ylabel('Salary')
plt.show()
```
